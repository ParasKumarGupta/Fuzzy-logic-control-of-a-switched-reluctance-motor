```python
import numpy as np
import skfuzzy as fuzzy
from skfuzzy import control as ctrl
import matplotlib.pyplot as plt
import random as rd

current=ctrl.Antecedent(np.arange(0,21,1),'current')
current['NB'] = fuzzy.trapmf(current.universe,[0,0,2,5])
current['NM'] = fuzzy.trimf(current.universe,[2,5,8])
current['NS'] = fuzzy.trimf(current.universe,[5,8,11])
current['Z'] = fuzzy.trimf(current.universe,[8,11,14])
current['PS'] = fuzzy.trimf(current.universe,[11,14,17])
current['PM'] = fuzzy.trimf(current.universe,[14,17,20])
current['PB'] = fuzzy.trimf(current.universe,[17,20,20])
current.view()
plt.show()
```

```python
position=ctrl.Antecedent(np.arange(0,51,5),'position')
position['NB'] = fuzzy.trapmf(position.universe,[0,0,5,10])
position['NM'] = fuzzy.trimf(position.universe,[5,10,15])
position['NS'] = fuzzy.trimf(position.universe,[10,15,20])
position['Z'] = fuzzy.trimf(position.universe,[15,25,30])
position['PS'] = fuzzy.trimf(position.universe,[25,30,35])
position['PM'] = fuzzy.trimf(position.universe,[30,35,40])
position['PB'] = fuzzy.trapmf(position.universe,[35,40,50,50])
position.view()
plt.show()
```

```python
compensated_current=ctrl.Consequent(np.arange(0,4,0.2),'compensated_current')
compensated_current['Z'] = fuzzy.trapmf(compensated_current.universe,[0,0,0.2,0.4])
compensated_current['PS'] = fuzzy.trimf(compensated_current.universe,[0.2,0.4,1])
compensated_current['PM'] = fuzzy.trimf(compensated_current.universe,[0.6,1.2,1.6])
compensated_current['PB'] = fuzzy.trimf(compensated_current.universe,[1.4,1.8,2])
compensated_current['PVB'] = fuzzy.trimf(compensated_current.universe,[2,2,3])
compensated_current['PVVB'] = fuzzy.trapmf(compensated_current.universe,[2.4,3,5,5])
compensated_current.view()
plt.show()
```

```python
rule1=ctrl.Rule(current['NB'] & position['NB'],compensated_current['PVVB'])
rule2=ctrl.Rule(current['NB'] & position['NM'],compensated_current['PVVB'])
rule3=ctrl.Rule(current['NB'] & position['NS'],compensated_current['PVB'])
rule4=ctrl.Rule(current['NB'] & position['Z'],compensated_current['PM'])
rule5=ctrl.Rule(current['NB'] & position['PS'],compensated_current['PB'])
rule6=ctrl.Rule(current['NB'] & position['PM'],compensated_current['PVB'])
rule7=ctrl.Rule(current['NB'] & position['PB'],compensated_current['PVVB'])
rule8=ctrl.Rule(current['NM'] & position['NB'],compensated_current['PVVB'])
rule9=ctrl.Rule(current['NM'] & position['NM'],compensated_current['PVVB'])
rule10=ctrl.Rule(current['NM'] & position['NS'],compensated_current['PVB'])
rule11=ctrl.Rule(current['NM'] & position['Z'],compensated_current['PS'])
rule12=ctrl.Rule(current['NM'] & position['PS'],compensated_current['PB'])
rule13=ctrl.Rule(current['NM'] & position['PM'],compensated_current['PVB'])
rule14=ctrl.Rule(current['NM'] & position['PB'],compensated_current['PVVB'])
rule15=ctrl.Rule(current['NS'] & position['NB'],compensated_current['PVVB'])
rule16=ctrl.Rule(current['NS'] & position['NM'],compensated_current['PVB'])
rule17=ctrl.Rule(current['NS'] & position['NS'],compensated_current['PB'])
rule18=ctrl.Rule(current['NS'] & position['Z'],compensated_current['PS'])
rule19=ctrl.Rule(current['NS'] & position['PS'],compensated_current['PM'])
rule20=ctrl.Rule(current['NS'] & position['PM'],compensated_current['PB'])
rule21=ctrl.Rule(current['NS'] & position['PB'],compensated_current['PVVB'])
rule22=ctrl.Rule(current['Z'] & position['NB'],compensated_current['PVVB'])
rule23=ctrl.Rule(current['Z'] & position['NM'],compensated_current['PB'])
rule24=ctrl.Rule(current['Z'] & position['NS'],compensated_current['PM'])
rule25=ctrl.Rule(current['Z'] & position['Z'],compensated_current['Z'])
rule26=ctrl.Rule(current['Z'] & position['PS'],compensated_current['PM'])
rule27=ctrl.Rule(current['Z'] & position['PM'],compensated_current['PB'])
rule28=ctrl.Rule(current['Z'] & position['PB'],compensated_current['PVVB'])
rule29=ctrl.Rule(current['PS'] & position['NB'],compensated_current['PVB'])
rule30=ctrl.Rule(current['PS'] & position['NM'],compensated_current['PB'])
rule31=ctrl.Rule(current['PS'] & position['NS'],compensated_current['PM'])
rule32=ctrl.Rule(current['PS'] & position['Z'],compensated_current['Z'])
rule33=ctrl.Rule(current['PS'] & position['PS'],compensated_current['PM'])
rule34=ctrl.Rule(current['PS'] & position['PM'],compensated_current['PB'])
rule35=ctrl.Rule(current['PS'] & position['PB'],compensated_current['PVB'])
rule36=ctrl.Rule(current['PM'] & position['NB'],compensated_current['PB'])
rule37=ctrl.Rule(current['PM'] & position['NM'],compensated_current['PS'])
rule38=ctrl.Rule(current['PM'] & position['NS'],compensated_current['PS'])
rule39=ctrl.Rule(current['PM'] & position['Z'],compensated_current['Z'])
rule40=ctrl.Rule(current['PM'] & position['PS'],compensated_current['PS'])
rule41=ctrl.Rule(current['PM'] & position['PM'],compensated_current['PM'])
rule42=ctrl.Rule(current['PM'] & position['PB'],compensated_current['PB'])
rule43=ctrl.Rule(current['PB'] & position['NB'],compensated_current['PM'])
rule44=ctrl.Rule(current['PB'] & position['NM'],compensated_current['PS'])
rule45=ctrl.Rule(current['PB'] & position['NS'],compensated_current['PS'])
rule46=ctrl.Rule(current['PB'] & position['Z'],compensated_current['Z'])
rule47=ctrl.Rule(current['PB'] & position['PS'],compensated_current['PS'])
rule48=ctrl.Rule(current['PB'] & position['PM'],compensated_current['PM'])
rule49=ctrl.Rule(current['PB'] & position['PB'],compensated_current['PB'])
rule=[rule1,rule2,rule3,rule4,rule5,rule6,rule7,rule8,rule9,rule10,rule11,rule12,rule13,rule14,rule15,rule16,rule17,rule18,rule19,rule20,rule21,rule22,rule23,rule24,rule25,rule26,rule27,rule28,rule29,rule30,rule31,rule32,rule33,rule34,rule35,rule36,rule37,rule38,rule39,rule40,rule41,rule42,rule43,rule44,rule45,rule46,rule47,rule48,rule49]
```

```python
a=ctrl.ControlSystem(rule)
b=ctrl.ControlSystemSimulation(a)
```

```python
b.input['current'] = rd.randint(0,20)
b.input['position'] = rd.randint(0,50)
b.compute()
print(b.output['compensated_current'])
```

```python
compensated_current.view(sim=b)
plt.show()
```
